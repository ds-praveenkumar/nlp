# Load Data

```python
df_dict = {}
df_all = pd.DataFrame()
data_dir = Path( '/kaggle/input/hacker-earth-ds-hiring/dataset/')
def read_data(data_dir: Path):
    for file in data_dir.glob("*.csv"):
        df_name = str(file.name).split('.')[0]
        df_ = pd.read_csv(file)
        df_dict[df_name] = df_
    print(f"No. of files read: {len(df_dict)}")
read_data( data_dir)
```

```python
#files name
df_dict.keys()
```

```python
df_sub = df_dict.pop('sample_submission')

def get_nunique( df_dict= df_dict):
    global df_all
    for key in df_dict.keys():
        print(key, "No. of columns:",  len( df_dict[key].columns ) )
        df_dict[key]['label'] = key
        df_all = df_all.append( df_dict[key] , ignore_index=True)
    print( "total rows:", df_all.shape[0])
get_nunique()
```

```python
def fix_columnnames(df_all =df_all):
    columns = [ str(col).replace(' ', '_').replace( '(', '').replace(')', '')
               for col in df_all.columns.to_list()]
    df_all.columns = columns
fix_columnnames()
```

```python
def get_unique_count( df_all= df_all):
    train = df_all[ df_all['label'] == 'train'] 
    test = df_all[df_all['label'] == 'test']
    for col in train.columns.to_list():
        print( f"train {col} unique values--> {train[col].nunique()} || test {col} unique values--> {test[col].nunique()}")
get_unique_count()
```
